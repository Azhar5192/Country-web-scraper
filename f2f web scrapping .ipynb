```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=options)

wait = WebDriverWait(driver, 20)  # wait up to 20 sec

driver.get("https://www.friends2support.org/inner/news/searchresult.aspx")

# Wait for Blood Group dropdown
blood = wait.until(
    EC.presence_of_element_located((By.ID, "dbBloodGroup"))
)

# Select values
Select(blood).select_by_visible_text("B+")
Select(wait.until(EC.presence_of_element_located((By.ID, "dbCountry"))))\
    .select_by_visible_text("INDIA")

Select(wait.until(EC.presence_of_element_located((By.ID, "dbState"))))\
    .select_by_visible_text("Bihar")

# Click Search
search_btn = wait.until(
    EC.element_to_be_clickable((By.ID, "btnSearch"))
)
search_btn.click()

# Wait for results table
table = wait.until(
    EC.presence_of_element_located((By.ID, "dgBloodDonorResults"))
)

rows = table.find_elements(By.TAG_NAME, "tr")

print("Name | Status | Mobile")
print("-" * 40)

for row in rows[1:]:
    cols = row.find_elements(By.TAG_NAME, "td")

    name = cols[0].text.strip()
    status = cols[1].text.strip()
    mobile = cols[2].text.strip()

    print(name, "|", status, "|", mobile)

driver.quit()
```
