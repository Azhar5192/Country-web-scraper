```python
import requests
import csv
from bs4 import BeautifulSoup 

req = requests.get("https://www.codewithharry.com/")

with open("data/index.html" , "w") as f:

    req.status_code
```

```python

carlist = []

# Read HTML file
with open("data/index.html", encoding="utf-8") as f:

    soup = BeautifulSoup(f.read(), "html.parser")


divs = soup.find_all('div', class_="o-fb o-gr o-eD o-fT o-ek DMTaiM o-lo o-ne o-nn")

for d in divs:
    car_name = d.find("div" , class_="o-o aXlX4B o-hl o-iT o-jq o-j3 o-jj o-bM o-cZ o-ci o-dv o-cA o-dN o-c4 o-dh o-jH").get_text(strip=True)
    car_price = d.find("div" , class_="o-j2 o-ji o-jJ  o-jq o-kJ o-aF o-f").get_text(strip=True)
    carlist([car_name,car_price])
    # print(car_name)

with open("car.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["Brands","Price"])
    writer.writerows(carlist)


print("file ready")
```
